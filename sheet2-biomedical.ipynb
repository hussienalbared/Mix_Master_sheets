```python
def expected_value(n):
  assert n>=1
  i_s=np.array([i for i in range(1,n+1)])
  m_n=i_s.sum()
  p_n=np.array([(i/m_n) for i in range(1,n+1)])
  return (i_s*p_n).sum()
```

```python
expected_value(100)
```

```python
def var(n):
  assert n>=1
  i_s=np.array([i for i in range(1,n+1)])
  i_s_2=np.array([i**2 for i in range(1,n+1)])
  m_n=i_s.sum()
  p_n=np.array([(i/m_n) for i in range(1,n+1)])
  ex_value=expected_value(n)
  return (i_s_2*p_n).sum()-ex_value**2
```

```python
var(100)
```
